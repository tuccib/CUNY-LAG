```python
missing_values = ["n/a", "na", "--"]
df3 = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/property.csv', na_values = missing_values)
df3
```

```python
df3.dtypes
```

```python
df3['PID'] = pd.to_numeric(df3['PID'], errors = 'coerce')
df3['PID'].fillna("10000001", inplace=True)
df3['ST_NUM'].fillna("200", inplace=True)
df3['OWN_OCCUPIED'].fillna("y", inplace=True)
# median_bedrooms = df['NUM_BEDROOMS'].median()
  # can't apply median if column has Nan
df3['NUM_BEDROOMS'].fillna(5, inplace=True)
df3['NUM_BATH'].fillna("4", inplace=True)
# mean_sqft = df['SQ_FT'].mean()
df3['SQ_FT'].fillna(555, inplace=True)
df3
```
